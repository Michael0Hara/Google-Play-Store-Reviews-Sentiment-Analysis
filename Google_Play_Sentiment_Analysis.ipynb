```python
import pandas as pd
import numpy as np
from google_play_scraper import app
from google_play_scraper import reviews_all
from google_play_scraper import Sort
from app_store_scraper import AppStore
import seaborn as sns
```

**Extraction of reviews from app- Amazon Mini TV**

```python
kuv = reviews_all("com.gooogle.android.kuvera.app",
                   sleep_milliseconds = 0,
                   lang='en',
                   country = 'us',
                   sort = Sort.NEWEST,
      )
```

```python
kuv_df = pd.json_normalize(kuv)
```

```python
kuv_df.head()
```

```python
kuv_df['score'].mean()
```

```python
kuv_df['reviewCreatedVersion'].value_counts()
```

**Transformers Model Initialization**

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")

print(sentiment_analysis("I love this!"))
```

```python
kuv_df.dtypes
```

```python
kuv_df['content'] = kuv_df['content'].astype('str')
```

```python
kuv_df['results'] = kuv_df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
kuv_df.head()
```

```python
kuv_df.shape
```

```python
kuv_df['sentiment'] = kuv_df['results'].apply(lambda x: (x[0]['label']))
kuv_df['s_score'] = kuv_df['results'].apply(lambda x: (x[0]['score']))
```
